```python
#Moyenne des dons
df['avg_pledged'].fillna(0, inplace=True)
df.replace(to_replace=[np.inf, -np.inf], value=0, inplace=True)
print('Les backers donnent en moyenne {} dollars pour une campagne'.format(round(df['avg_pledged'].mean(), 2)))
```

```python
category_list = list(df['main_category'].unique())

pledge_cat = dict()
for category in category_list:
    category_df = df[df['main_category'] == category]
    pledge_cat.update({category : category_df['avg_pledged'].mean()})
    
# Visualize the results
pledge_df = pd.DataFrame(pledge_cat, index=['avg_pledged'])
pledge_df.sort_values('avg_pledged', axis=1, inplace=True, ascending=False)
ax = pledge_df.plot(kind='bar', 
                    title='Average pledged per backer by category', 
                    rot=0, figsize=(20,5))
ax.set_xlabel('Categories')
ax.set_ylabel('Average pledged per backer (USD)')
```

Ce sont donc les projets concernant les technologies, l'élaboration du design d'un produit et les films qui bénéficient des montants par donneur les plus élevés.

## Preprocessing time variables

```python
df['launched'].head()
```

```python
from datetime import date
from datetime import time
from datetime import datetime
import datetime as dt
```

```python
#Mise en forme de la variable de la date de lancement
df['launched'] = df['launched'].apply(lambda x: dt.datetime.strptime(x,"%Y-%m-%d %H:%M:%S"))
df['launched'].head()
```

```python
#comparaison des erreurs numériquement
from sklearn.metrics import mean_squared_error
err = mean_squared_error(y_test, predict)
err_log = mean_squared_error(np.log(y_test+1), predict_log)
print(err)
print(err_log)
```

Le modèle linéaire malheureusement a ses failles, l'erreur quadratique moyenne est énorme. C'est du aux valeurs extrêmes que le modèle n'arrive pas du tout à prédire. Il est d'ailleurs systématiquement plus optimiste que la réalité si l'on regarde le graph des erreurs qui se trouve plus bas. 
A l'inverse l'erreur quadratique moyenne du modèle log-linéaire paraît beaucoup plus réduite mais la valeur est exprimée en log. En réalité l'erreur quadratique moyenne est de 22 000 dollars. 

```python
#comparaison des erreurs graphiquement
#plot erreurs pour le modèle linéaire
fig = plt.figure(figsize=(5, 5))
ax = plt.subplot()
ax.plot(y_test, predict, '.')
ax.set_xlabel("expected")
ax.set_ylabel("predicted")
ax.set_title("ERR2={0}".format(err))

#plot erreurs pour le modèle log linéaire
fig = plt.figure(figsize=(5, 5))
ax = plt.subplot()
ax.plot(y_test, predict_log, '.')
ax.set_xlabel("expected")
ax.set_ylabel("predicted")
ax.set_title("ERR2={0}".format(err_log))

#Comparaison des erreurs triées par ordre croissant
y1 = list(sorted(predict - y_test))
y2 = list(sorted(predict_log - np.log(y_test+1)))
figcomp = plt.figure(figsize=(5, 5))
ax = plt.subplot()
ax.plot([0, len(y1)], [0, 0], '-')
ax.plot(y1, '.', label="linear")
ax.plot(y2, '.', label="log linear")
ax.legend()
plt.show()
```

La projection des points expected/predicted est très concentrée dans le coin en bas à gauche car la grande majorité des données correspond à des petits montants. Néanmoins, certains points 'outliers' concernent des montants très importants. 

La prédiction log linéaire apparaît clairement moins pertinente que la régression linéaire simple car elle n'arrive pas à prédire les campagnes qui ne récoltent rien. 
La régression linéaire simple n'est pas parfaite car elle ne forme pas une diagonale, et se trouve être en forme de parapluie. Jusqu'à 1 million de dollars, le modèle est plutôt performant. En revanche, pour des valeurs plus élevées, il n'arrive plus à prédire, ce qui explique que notre erreur moyenne quadratique est très forte. 

Ceci peut s'expliquer par plusieurs facteurs : 

1- les données pour des sommes aussi élevées sont très peu nombreuses dans notre dataset

2- les montants récoltés sont souvent décorrelés des variables qui conduisent au succès. Si l'on prend l'exemple de la campagne qui a récolté le plus de fonds, le montant récolté est beaucoup plus important que le montant demandé en objectif, plus de 7 000 personnes ont participé en donnant 250 dollars en moyenne (chiffres beaucoup plus hauts que la moyenne du dataset). De plus, il s'agissait de fonds pour une technologie dont le nom ne contient pas les mots clés "porteurs de succès". 

Concernant le graph des résidus : 

1- pour le modèle linéaire, les résidus ne suivent pas une distribution normale. La courbe est en forme de S inversé avec des queues de distribution très importantes

2- Les résidus du modèle log-linéaire ne suivent pas non plus une distribution normale mais il n'y a pas de queues de distribution du tout. 
Ainsi, il peut y avoir de l'hétéroscedasticité dans le modèle mais cela n'empêche pas les estimateurs OLS d'être non-biaisés et consistents dans des larges dataset. 

Face à ce constat, nous allons essayer de trouver un  modèle qui prédit mieux les valeurs plus élevées.

## Random Forest Classification

Nous avons voulu comparer entre une arbre de décision et le random forrest et trouver le bon nombre d'estimations (pour la RF)  à partir duquel on améliore les prédictions son tomber non plus dans la "sur-estimation". 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
n_estimators = [1, 10, 25, 50, 100]
train_results = []
test_results = []
for estimator in n_estimators:
   rf = RandomForestClassifier(n_estimators=estimator, n_jobs=-1)
   rf.fit(X_train, y_train)
   train_pred = rf.predict(X_train)
   false_positive_rate, true_positive_rate, thresholds = roc_curve(y_train, train_pred)
   roc_auc = auc(false_positive_rate, true_positive_rate)
   train_results.append(roc_auc)
   y_pred = rf.predict(X_test)
   false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
   roc_auc = auc(false_positive_rate, true_positive_rate)
   test_results.append(roc_auc)
from matplotlib.legend_handler import HandlerLine2D
line1, = plt.plot(n_estimators, train_results, 'b', label="Train AUC")
line2, = plt.plot(n_estimators, test_results, 'r', label="Test AUC")
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.ylabel('AUC score')
plt.xlabel('n_estimators')
plt.show()
```

Ainsi, nous pouvons voir qu'utiliser plus de 50 n_estimators ne rajoute rien à la performance de la classification. Nous aurions aimé jouer sur d'autres paramètres du modèle (comme max_depth) mais ce fut compliquer en raison de limitation de RAM.

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=50)
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

Les meilleurs résultats sont obtenus avec le RandomForestClassifier. On reste cependant louin d'une AUC proche de 1.
